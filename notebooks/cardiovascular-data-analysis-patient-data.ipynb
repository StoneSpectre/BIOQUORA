## About Dataset

The dataset was originally created by SVETLAN_ULIANOVA* and this is a public kernelin kaggle

The dataset consists of 70 000 records of patients data in 12 features, such as age, gender, systolic blood pressure, diastolic blood pressure, and etc. 

To visit the original page of posting of the dataset on kaagle click here [link](https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset)

#### Data description

There are 3 types of input features:

Objective: factual information;
Examination: results of medical examination;
Subjective: information given by the patient.

#### Features

* Age | Objective Feature | age | int (days)
* Height | Objective Feature | height | int (cm) |
* Weight | Objective Feature | weight | float (kg) |
* Gender | Objective Feature | gender | categorical code |
* Systolic blood pressure | Examination Feature | ap_hi | int |
* Diastolic blood pressure | Examination Feature | ap_lo | int |
* Cholesterol | Examination Feature | cholesterol | 1: normal, 2: above normal, 3: well above normal |
* Glucose | Examination Feature | gluc | 1: normal, 2: above normal, 3: well above normal |
* Smoking | Subjective Feature | smoke | binary |
* Alcohol intake | Subjective Feature | alco | binary |
* Physical activity | Subjective Feature | active | binary |
* Presence or absence of cardiovascular disease | Target Variable | cardio | binary |
  All of the dataset values were collected at the moment of medical examination.
  
## Data Cleaning
#### Prepare and processs the data

* Packages installed are : tidyverse,diplyr,tidyr
```{r echo=TRUE}
install.packages("tidyverse")
library("tidyverse")
install.packages("dplyr")
library("dplyr")
install.packages("tidyr")
library("tidyr")
```

* To import and read the file containing the dataset : 
```{r echo=TRUE, results='hide'}
library(readr)
cardio_train<-read.table(file="cardio_train.csv", header = TRUE, sep = ';') #here the sep= ";" is used because the original file was in csv format and the delimeter used to seperate the columns and values is ";" a semicolon 
```

* The format for the age in column "age" was in days, Converting to the format in years:
```{r echo=TRUE, results='hide'}
cardio_train1<-mutate(cardio_train,age=age%/%365)#assigned to new datatable "cardio_train1"
```

* Arrange the table based on the age of the recorded population in ascending order :
```{r echo=TRUE, results='hide'}
cardio2<-arrange(cardio_train1, age)#assigned to new datatable "cardio2"
```

* Converted the values of columns (alco, smoke, active, cardio) which has "int" type values 1 and 0 to "YES" and "NO" respectivley
```{r echo=TRUE, results='hide'}
cardio2$smoke[cardio2$smoke==0] <- "NO"
cardio2$smoke[cardio2$smoke==1] <- "YES"
cardio2$alco[cardio2$alco==0]<-"NO"
cardio2$alco[cardio2$alco==1]<-"YES"
cardio2$active[cardio2$active==0]<-"NO"
cardio2$active[cardio2$active==1]<-"YES"
cardio2$cardio[cardio2$cardio==0]<-"NO"
cardio2$cardio[cardio2$cardio==1]<-"YES"
```

* Converted the values of the column gender from 1 and 2 (int type) to "Female" and "Male"(string type) respectivley and based on the dataset
```{r echo=TRUE, results='hide'}
cardio2$gender[cardio2$gender==1] <- "Female"
cardio2$gender[cardio2$gender==2] <- "Male"
```

* To remove any blank spaces or NULL values
```{r echo=TRUE, results='hide'}
drop_na(cardio2)
```

* As some of the inputs or reported vlues i the dataset were incorrect
Lowest height and lowest height were 50cm and 250cm
Lowest weight and lowest weight were 10kg and 200kg
Lowest systolic_BP and highest systolic_BP -150 and 16020 ......etc
All the above values are not normal and authentic
* Filtering out data was necessary to get accurate
```{r echo=TRUE, results='hide'}
cardio2_fil<- cardio2 %>% 
filter(`ap_hi`<180 & `ap_hi`>50, `ap_lo`>35 & `ap_lo`<110, height<220 & height>140, weight>50 & weight<160)
```
Filtering was done and the limits were set based on the statistics of viable human being that can survive in extreme cases

* Rename the columns names : some columns names didn't gave what context it has, for beter understanding
```{r echo=TRUE, results='hide'}
colnames(cardio2_fil)[6]<-"systolic_BP"
colnames(cardio2_fil)[7]<-"diastolic_BP"
colnames(cardio2_fil)[12]<-"Physical_activity"
colnames(cardio2_fil)[13]<-"Cardio_vascular_disease"
```

**Before data cleaning**

[![before1.png](https://i.postimg.cc/8cHkjqZ9/before1.png)](https://postimg.cc/K31hWsSD)

**After data cleaning**

[![after1.png](https://i.postimg.cc/hvgP4S1V/after1.png)](https://postimg.cc/9R8jJHSf)


* A Short summary of the dataset: Shows a report of how likely is each gender prone to cardiovascular disease

```{r echo=TRUE}
summ_2<-cardio2_fil %>% 
group_by(gender) %>% 
summarise(avg_systolic_BP=mean(`systolic_BP`), avg_diastolic_BP=mean(`diastolic_BP`), percentage_likly=sum(`Cardio_vascular_disease`=="YES")/n(), count=n(),.groups = "drop") 
view(summ_2)
```

## Share and Act

#### Insights

* Systolic Blood pressure higher than 120mmHg are more prone to cardiovasular disease for both male and female
  (MAX for both gender= 140-150 mmHg)
* Diastolic Blood pressure higher than 80mmHg are prone to cardiovascular disease for females as for males above 90mmHg are prone to the disease
  (MAX for female=80mmHg, MAX for male=100mmHg)
* Person with glucose level 3 is more likely to get cardiovascular disease for both male and female 
  (MAX= glucose level 3)
* Person with cholesterol level 3 is more likely to get cardiovascular disease for both male and female
  (MAX = cholesterol level 3)
* Physical activity or a person being physically active has no significant effect on glucose and cholesterol level for both male and female

#### Suggestions

* Lose extra kg of weights and watch the waistline to reduce systolic and diastolic BP,
  Men are at risk if their waist measurement is greater than 40 inches(102cm)
  Women are at risk if their waist measurement is greater than 35 inches(89cm)
* Eat healthy diet, low saturated fat and low cholesterol can reduce blood pressure by upto 11mmHg
* Reduce salt/sodium in the diet, eat fewer processed foods
* Quit smoking and limit alcohol consumption, females who smoke and drink are more affected by cardiovascular disease
* Try acupuncture - helps to reduce blood pressure in people who takes medication to reduce blood pressure
*Eating too much such as snacking between meals can cause high blood sugar/glucose level
*Prevent dehydration(dehydration causes hyperglycemia), it it suggested to drink at least 4 liters of water per day for Men and 2.7 liters for Women
